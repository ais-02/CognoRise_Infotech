```python
df1 = pd.get_dummies(df,columns = ['CarName','fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem'],dtype = int)
change = {'two':2,'four':4}
df1['doornumber'] = df['doornumber'].map(change)
change2 = {'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12}
df1['cylindernumber'] = df['cylindernumber'].map(change2)
df1
```

```python
x = df1.drop(['price'],axis = 1)
y = df1['price']
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_percentage_error
```

```python
md1 = KNeighborsRegressor()
md2 = LinearRegression()
md3 = RandomForestRegressor()
md4 = DecisionTreeRegressor()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
```

```python
list1 = [md1,md2,md3,md4]
```

```python
for i in list1:
  print('model started',i)
  i.fit(x_train,y_train)
  y_pred = i.predict(x_test)
  print('r2 score : ',r2_score(y_test,y_pred))
  print('% error : ',mean_absolute_percentage_error(y_test,y_pred))
  print('model ended')
  print()
```
